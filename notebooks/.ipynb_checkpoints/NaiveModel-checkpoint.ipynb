# Naive Model - true baseline

We consider if classic: $y_t = y_{t-1}$ model might be reasonable in such a hard problem!
Let's create baseline predictions for further models!

### Dependencies loading

```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
import warnings
warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', 500)
pd.set_option('display.max_rows', 150)
```

```python
df = pd.read_csv("../data/train_fe.csv", index_col=0)
df_test = pd.read_csv("../data/test_fe.csv",index_col=0)
```

### Dataset preparation

```python
X = df[["Ticker","etr","rok"]]
```

```python
Y = df_test[["Ticker","etr","rok"]]
```

```python
y_pred = X[X.rok == 2015].etr.tolist()
```

```python
y_true = Y.etr.tolist()
```

```python
rmse = np.sqrt(mean_squared_error(y_true, y_pred))
```

```python
mae = mean_absolute_error(y_true, y_pred)
```

```python
pd.DataFrame([rmse, mae]).T.rename(columns ={0:"RMSE", 1:"MAE"})
```

### Saving final predictions

```python
pd.DataFrame([rmse, mae]).T.rename(columns ={0:"RMSE", 1:"MAE"}).to_csv("final_models/naive.csv")
```
